```python
#Slice "Artificial" to get "Art" and "ficial" seperately
word = "Artificial"
word1 = word[0:3]
word2 = word[4:10]
print(word1)
print(word2)
```

```python
#Given word = "PythonPrograaming", extract "Python" and "Programming" using slicing

word = "PythonProgramming"
word1 = word[0:6]
word2 = word[6:17]
print(word1)
print(word2)
```
